2. Importación de librerías

```python
import chromadb
from sentence_transformers import SentenceTransformer
```

3. Inicialización de la base de datos y modelo de embedding

```python
# Crear cliente y colección
client = chromadb.Client()
collection = client.create_collection(name="Tecsup")

# Usamos un modelo de embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
```

 4. Información personal y de compañeros (ejemplo)

```python
docs = [
    "Deivis es estudiante de Big Data en Tecsup",
    "Carlos estudia Ciberseguridad y le interesa la minería de textos",
    "Fiorella participa en proyectos de inteligencia artificial",
    "Bryan es experto en visualización con Power BI",
    "Lucía está desarrollando una app para la gestión educativa"
]

embeddings = model.encode(docs).tolist()

collection.add(
    documents=docs,
    embeddings=embeddings,
    ids=["1", "2", "3", "4", "5"]
)
```

5. Realiza una búsqueda en la base vectorial

```python
query = "¿Quién trabaja con inteligencia artificial?"
query_embedding = model.encode(query).tolist()

results = collection.query(
    query_embeddings=query_embedding,
    n_results=2
)

print(results['documents'])
```

 6. Visualización de embeddings

```python
data = collection.get(include=["embeddings", "documents"])

for doc, _id, emb in zip(data['documents'], data['ids'], data['embeddings']):
    print(f"ID: {_id}")
    print(f"Documento: {doc}")
    print(f"Embedding (primeras 5 dimensiones): {emb[:5]}")
    print("-" * 30)
```
